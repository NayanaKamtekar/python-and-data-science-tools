```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('responses.csv', delimiter = ",") # read in the data and print it
df # Note that the first line of the csv-file is interpreted as variable names (which fits in this case!). 
```

```python
df.dtypes
```

```python
df['min'] = pd.to_numeric(df['min'], errors = 'coerce') # change min from string to numeric
df
```

```python
df['time'] = df['min']*60 + df.sec # make a time variable
df
```

#### Summaries
Summary statistics can help us understand the data (and also clean it).

```python
df.shape # size of data (number of (rows,columns) - recall, Python is zero-indexed)
```

#### Data cleaning

```python
df.loc[~df['exp'].isin(['Yes','No']),'exp'] = np.nan # if cor and exp variable has yes,no values then replaced by NaN
df.loc[~df['cor'].isin(['Yes','No']),'cor'] = np.nan
df
```

```python
df.loc[df['sec'] > 60,'sec'] = np.nan # if sec variable has invalid values then replaced by NaN
df
```

```python
df1 = df.dropna()
df1
```

```python
df1['cor'].value_counts() # get counts for categorical data
```

#### Visualization
Visualization of data can help us understand it - which types of visualization that are meaningful will depend on the variables,

```python
import matplotlib.pyplot as plt
%matplotlib inline
# numeric data
df1.plot(x = 'Unnamed: 0', y = 'time', kind = 'scatter') # scatter plot 
df1.plot(y = 'time', kind = 'density') # density plot
```

```python
df1['type'].value_counts().plot(kind='bar') #visualisation of catagorical data
```

Investigate if the participants in the sudoku experiment were faster doing the Latin letter or the Greek letter sudoku by taking the mean time for each type of sudoku (using your clean data set).

```python
df1.groupby('type').agg({"time" :['mean']})
```

Mean time of Greek letter sudoku is less than mean time of latin letter sudoku. Therefore Greek letter sudoku is easier to solve.

### Task 3
Find other features that may be predictive of spam (for instance, as indicated by histograms).

```python
email_df = pd.read_csv('SMSSpamCollection.txt', delimiter='\t', names=['s', 't'])
email_df['len'] = email_df['t'].str.len()
email_df.head(50)
```

### The spam emails have longer text
Summary of length of email text shows that on average the spam emails are longer. 
The legth of text has different distribution for spam.

```python
email_df.groupby('s').agg(['count', 'median', 'mean'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
email_df.hist(column = 'len', bins = 50, by = 's')
```

### Spam emails contains post box numbers or mobile telephone numbers
Around 80% of spam emails contain a string of number more than length 5, which can be either post box numbers or mobile numbers. Compared to this only 0.08% ham emails contain such a number string.

```python
email_df['num_seq_exist'] = email_df['t'].str.replace(' ','').str.match('(.*)(\d{5,})(.*)')
email_df[['num_seq_exist', 's']].groupby('s').sum() / email_df[['num_seq_exist', 's']].groupby('s').count() * 100
```

### Distribution of numbers in the text
The distribution of count of numbers in the text differes between ham and spam emails

```python
email_df['num_count']= email_df.t.str.count('\d')
```
